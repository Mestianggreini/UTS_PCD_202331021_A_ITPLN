## Import Library

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Loding Gambar

```python
image_path = 'img_orang.jpg'  # Pastikan gambar ada di folder yang sama
image = cv2.imread(image_path)
```

## Cek jika gambar gagal dimuat

```python
if image is None:
    raise FileNotFoundError(f"Gambar '{img_orang}' tidak ditemukan.")
```

## Konversi BGR (OpenCV) ke RGB (matplotlib)

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

## Konversi ke grayscale

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

## Histogram Equalization

```python
hist_eq = cv2.equalizeHist(gray)
```

## Manual Contrast Stretching

```python
# Manual brightness + contrast adjustment (dipencerah dan diperkontras)
alpha = 1.5  # Kontras (1.0 = normal)
beta = 40    # Kecerahan (0 = normal)
bright_contrast = cv2.convertScaleAbs(gray, alpha=alpha, beta=beta)
```
